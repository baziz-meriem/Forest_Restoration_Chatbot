```python
import os
from PyPDF2 import PdfReader
```

### Text extracting function

```python
def extract_text_from_pdf(pdf_path):
    pdfreader = PdfReader(pdf_path)
    raw_text = ''
    for i, page in enumerate(pdfreader.pages):
        content = page.extract_text()
        if content:
            raw_text += content
    return raw_text
```

```python
# Main directory containing all the folders
main_directory = 'Module_1_Fundamentals_of_Ecological_Restoration-20230913T195003Z-001/Module_1_Fundamentals_of_Ecological_Restoration'
outside_directory = "Module_1_Fundamentals_of_Ecological_Restoration-20230913T195003Z-001"

# Create a function to ensure the destination directory exists
def ensure_directory_exists(directory_path):
    if not os.path.exists(directory_path):
        os.makedirs(directory_path)

# Iterate over each folder in the main directory
for folder_name in os.listdir(main_directory):
    folder_path = os.path.join(main_directory, folder_name)
    print(f"Extracting from folder:{folder_name}")
    print("--------------------------------------")
    # Check if it's a directory
    if os.path.isdir(folder_path):
        combined_text = ''

        # Iterate over each file in the folder
        for file_name in os.listdir(folder_path):
            if file_name.endswith('.pdf'):
                pdf_path = os.path.join(folder_path, file_name)
                print(f"Extracting pdf:{file_name}")
                combined_text += extract_text_from_pdf(pdf_path)

         # Define the destination directory and ensure it exists
        dest_directory = os.path.join(outside_directory+ "_txt")
        ensure_directory_exists(dest_directory)

         # Save the combined text to a .txt file named after the folder
        with open(os.path.join(dest_directory, folder_name + '.txt'), 'w',encoding='utf-8') as txt_file:
            txt_file.write(combined_text)
```

## Final Code
[pyPDFLoader](https://python.langchain.com/docs/modules/data_connection/document_loaders/pdf)

[ref](https://stackoverflow.com/questions/76677150/python-based-way-to-extract-text-from-scientific-academic-paper-for-a-language-m)

```python
from langchain.document_loaders import PyPDFLoader
```

```python
# Create a function to ensure the destination directory exists
def ensure_directory_exists(directory_path):
    if not os.path.exists(directory_path):
        os.makedirs(directory_path)


def extract_text_with_langchain(pdf_path):
    loader = PyPDFLoader(pdf_path)
    pages = loader.load_and_split()
    text = ""
    for page in pages:
        text += page.page_content
        text += "\n"
    return text
```

```python
# Main directory containing all the folders
main_directory = 'Module_1_Fundamentals_of_Ecological_Restoration-20230913T195003Z-001/Module_1_Fundamentals_of_Ecological_Restoration'
outside_directory = "Module_1_Fundamentals_of_Ecological_Restoration-20230913T195003Z-001"

# Iterate over each folder in the main directory and extract text
for folder_name in os.listdir(main_directory):
    folder_path = os.path.join(main_directory, folder_name)
    print(f"Extracting from folder:{folder_name}")
    print("--------------------------------------")
    # Check if it's a directory
    if os.path.isdir(folder_path):
        combined_text = ''

        # Iterate over each file in the folder
        for file_name in os.listdir(folder_path):
            if file_name.endswith('.pdf'):
                pdf_path = os.path.join(folder_path, file_name)
                print(f"Extracting pdf:{file_name}")
                combined_text += extract_text_with_langchain(pdf_path)

        # Define the destination directory and ensure it exists
        dest_directory = os.path.join(outside_directory+"_txt_langchain")
        ensure_directory_exists(dest_directory)

        # Save the combined text to a .txt file named after the folder using UTF-8 encoding
        with open(os.path.join(dest_directory, folder_name + '.txt'), 'w', encoding='utf-8') as txt_file:
            txt_file.write(combined_text)
```
